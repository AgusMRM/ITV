```python
import matplotlib.patches as mpatches
import matplotlib.font_manager as font_manager
matplotlib.rc('xtick', labelsize=15)     
matplotlib.rc('ytick', labelsize=15)
axis_font = {'fontname':'Arial', 'size':'20'}
fig=plt.figure(figsize=(10,10))
fig.subplots_adjust(bottom=0.10, top=0.98, left=0.10, right=0.99)
vs=[]
vr=[]
plt.figure(figsize=(9,9))
for i in range(1357):
    x=str(i)
    t=np.loadtxt('/mnt/is2/fstasys/ITV/dpaz/Corrida_base09_voids/perfiles/void_'+x+'.dat')
    radio = t[0,4]
    rl3 = np.log10(3*radio)
    rl2 = np.log10(1.0*radio)
    abin = np.log10(t[99,0])/100.
    bn3 = int(rl3/abin)
    bn2 = int(rl2/abin)
    s   = t[bn2:bn3,3]
    cut, = np.where(s>0)
    #print(t[bn,3])
    
    #cut, = np.where(t[:,3]>0.8)
    #if (len(cut)>0):
    #    print(i)
    
    
    if (len(cut)>0):
        plt.plot(t[:,0]/radio,t[:,3],color='orangered',lw=.5)
        vs.append(t[0,4])
    else:
        plt.plot(t[:,0]/radio,t[:,3],color='turquoise',lw=.5)
        vr.append(t[0,4])       

plt.plot(t1[:,0]/t1[0,4],t1[:,3],color='black',lw=3, ls='dotted')        
plt.plot(t2[:,0]/t2[0,4],t2[:,3],color='black',lw=3, ls='dashed')        
plt.plot(t3[:,0]/t3[0,4],t3[:,3],color='black',lw=3, ls='dotted')        
plt.plot(t4[:,0]/t4[0,4],t4[:,3],color='black',lw=3, ls='dashed')        
plt.plot(t5[:,0]/t5[0,4],t5[:,3],color='black',lw=3, ls='dotted')        
plt.plot(t6[:,0]/t6[0,4],t6[:,3],color='black',lw=3, ls='dashed',label='s')        
plt.legend(loc='upper left')        
        
plt.semilogx()
red_patch = mpatches.Patch(color='orangered', label='Tipo S')
blue_patch = mpatches.Patch(color='turquoise', label='Tipo R')
#s_patch = mpatches.Patch(ls='dotted',color='black',label='S-selection')

plt.legend(handles=[red_patch,blue_patch],fontsize='xx-large',numpoints=3)
plt.xlim(.7,10)
plt.xlabel("$r/r_{void} $",size=25)
plt.ylabel(r"$\Delta(r)$",size=25)
plt.xticks([1,10],(1,10),size=20)
plt.minorticks_on()
#plt.savefig(fname='/home/arodriguez/Void-s-Gastrophysics/SeleccionGraficos/voids_sep.png',format='png')
```
