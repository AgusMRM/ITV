```python
t=np.loadtxt('PROM_b1000.100.dat')
dgs=(3*(100**2)*(0.045))/(8*np.pi*(4.3e-9))*1e-10 #Msol/Mpc3
t2=np.loadtxt('PROM_b100.10.dat')
t3=np.loadtxt('PROM_b500.10.dat')
```

```python
r0=np.loadtxt('particlesprofile_b50.dat')
r1=np.loadtxt('particlesprofile_b10.dat')
ddm=(3*(100**2)*(0.276-0.045))/(8*np.pi*(4.3e-9))*1e-10
```

```python
fig=plt.figure(1,figsize=(7,7))
plt.plot(t[:,0],t[:,1]/dgs-1,color='grey',label='gas sph 1000-100')
plt.plot(t2[:,0],t2[:,1]/dgs-1,color='black',label='gas sph 100-10')
plt.plot(t3[:,0],t3[:,1]/dgs-1,color='black',label='gas sph 500-10',ls='--')

plt.plot(r0[:,0],(r0[:,2]/r0[:,5]-ddm)/ddm,color='red',label='DM b50',lw=1)
plt.plot(r1[:,0],(r1[:,2]/r1[:,5]-ddm)/ddm,color='darkred',label='DM 10',lw=1,ls='--')

plt.xlim(3e-1,3e1)
plt.semilogx()
plt.legend()
plt.title('S void')
#plt.savefig(fname='/home/arodriguez/tesis/perfiles/S1373_sph2.pdf',format='pdf')
```

```python
t=np.loadtxt('profilegas_S.dat')
t2=np.loadtxt('particlesprofile_S.dat')

y=t[:,1]
dy=t[:,3]/dgs
ddm=(3*(100**2)*(0.276-0.045))/(8*np.pi*(4.3e-9))*1e-10
```

```python
dy_neg = (t2[:,1]-t2[:,3])/ddm -1
dy_pos = (t2[:,1]+t2[:,3])/ddm -1
```

```python
import matplotlib.font_manager as font_manager
matplotlib.rc('xtick', labelsize=15)     
matplotlib.rc('ytick', labelsize=15)
axis_font = {'fontname':'Arial', 'size':'20'}
fig=plt.figure(1,figsize=(9,7))

#plt.plot(t[:,0],t[:,1]/dgs-1,color='blue',label='jacknife')
plt.fill_between(t[:,0], (y - dy-dgs)/dgs, (y + dy-dgs)/dgs,color='blue', alpha=0.2)
plt.plot(t[:,0],t[:,1]/dgs-1,color='black',label='GAS',ls='--')

plt.plot(t2[:,0],(t2[:,1]-ddm)/ddm,color='red',label='DM',ls='--')
plt.fill_between(t2[:,0], dy_neg,dy_pos ,color='red', alpha=0.2)

plt.xlim(.4,24.5)
plt.legend()
plt.semilogx()
plt.xlabel("$radio [Mpc]$",size=20)
plt.ylabel(r"$\delta(r) [M_{\odot}/Mpc]$",size=20)

#plt.savefig(fname='/home/arodriguez/tesis/perfiles/R1198_sph3.png',format='png')
```
