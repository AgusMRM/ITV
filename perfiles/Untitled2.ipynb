```python
r0=np.loadtxt('S1373_resi.dat')
b=np.loadtxt('S1373_base.dat')
```

```python
ddm=(3*(100**2)*(0.276-0.045))/(8*np.pi*(4.3e-9))
dgs=(3*(100**2)*(0.045))/(8*np.pi*(4.3e-9))
ddm2=(3*(100**2)*(0.276))/(8*np.pi*(4.3e-9))

mp=ddm2*(500**3)/(512**3)
```

```python
mp
```

```python
plt.plot(r0[:,0],((r0[:,1])*1e10/r0[:,6]-dgs)/dgs,color='blue',label='GAS')#
plt.plot(r0[:,0],(r0[:,2]*1e10/r0[:,6]-ddm)/ddm,color='red',label='DM')
#plt.plot(r0[:,0],(r0[:,3]/r0[:,6]-dtd)/dtd,color='darkred',label='DM+TID')
#plt.plot(r0[:,0],(r0[:,4]/r0[:,6]-dst)/dst,color='orange',label='ESTRELLAS')
#plt.plot(r0[:,0],(r0[:,5]/r0[:,6]-dtt)/dtt,color='black',label='TOTAL')

plt.plot(b[:,0],((b[:,1]*mp/b[:,2])/ddm2 -1),color='grey',linestyle='--',label='BASE')
plt.legend()
plt.semilogx()
plt.title('DENSIDAD INTEGRADA S1373')
#plt.figure(figsize(8,8))
plt.savefig(fname='/home/arodriguez/tesis/perfiles/S1373_int.pdf',format='pdf')
```
