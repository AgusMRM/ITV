```python
dgs=(3*(100**2)*(0.045))/(8*np.pi*(4.3e-9))*1e-10 #Msol/Mpc3
ddm=(3*(100**2)*(0.276-0.045))/(8*np.pi*(4.3e-9))*1e-10
```

```python
t=np.loadtxt('profile_gas.dat')
y=t[:,1]
dy=t[:,3]/dgs
ddm=(3*(100**2)*(0.276-0.045))/(8*np.pi*(4.3e-9))*1e-10
```

```python
import matplotlib.font_manager as font_manager
matplotlib.rc('xtick', labelsize=15)     
matplotlib.rc('ytick', labelsize=15)
axis_font = {'fontname':'Arial', 'size':'20'}
fig=plt.figure(1,figsize=(9,7))

#plt.plot(t[:,0],t[:,1]/dgs-1,color='blue',label='jacknife')
plt.fill_between(t[:,0], (y - dy-dgs)/dgs, (y + dy-dgs)/dgs,color='blue', alpha=0.2)
plt.plot(t[:,0],t[:,1]/dgs-1,color='black',label='GAS',ls='--')


plt.xlim(.4,24.5)
plt.legend()
plt.semilogx()
plt.xlabel("$radio [Mpc]$",size=20)
plt.ylabel(r"$\delta(r) [M_{\odot}/Mpc]$",size=20)
```
