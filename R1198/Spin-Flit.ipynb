```python
t=np.loadtxt('halos_R1198.dat')
```

```python
xbox=411.2170 
ybox=162.1655
zbox=453.0553

x=413.621475  
y=162.604601 
z=448.953638

x = x - xbox + 250
y = y - ybox + 250
z = z - zbox + 250
```

```python
d=np.sqrt((t[:,8]-x)**2 + (t[:,9]-y)**2 + (t[:,10]-z)**2)
dx=t[:,8]-x
dy=t[:,9]-y
dz=t[:,10]-z
lx=t[:,14]
ly=t[:,15]
lz=t[:,16]
spn=t[:,18]
```

```python
L = np.sqrt(lx**2+ly**2+lz**2)
mask1, = np.where(L > 0)
tita = (lx[mask1]*dx[mask1] + ly[mask1]*dy[mask1] + lz[mask1]*dz[mask1])/(L[mask1]*d[mask1])
#tita=np.arccos(theta)
```

```python
grandes, = np.where( t[mask1,1] > 500)
chicos, = np.where( t[mask1,1]<50)
```

```python
plt.hist(tita,histtype='step',lw=3,color='black',bins=20,density='true',label='TODOS')
plt.hist(tita[grandes],histtype='step',linestyle='--',lw=2,color='red',density='true',label='grandes')
plt.hist(tita[chicos],histtype='step',linestyle='--',lw=2,color='blue',density='true',label='chicos')
plt.legend()
plt.xlabel('angulo')
plt.title('Spin-flit')
```

```python
plt.hexbin(d[chicos],tita[chicos],gridsize=(20,20))
plt.xlabel('distancia')
plt.colorbar()
plt.ylabel('angulo')
```
