```python
r0=np.loadtxt('partprof_R1198.dat')
s0=np.loadtxt('partprof_S1373.dat')
r1=np.loadtxt('partprof_R0832.dat')
r2=np.loadtxt('partprof_R1277.dat')
s1=np.loadtxt('partprof_S1050.dat')
```

```python
vol=(4./3.)*np.pi*(40.**3)
nr0=np.sum(r0[:,1])/vol
ns0=np.sum(s0[:,1])/vol
nr1=np.sum(r1[:,1])/vol
nr2=np.sum(r2[:,1])/vol
ns1=np.sum(s1[:,1])/vol
```

```python
plt.figure(figsize=(10,10))
plt.plot(r0[:,0],(r0[:,1]/r0[:,3]-nr0)/nr0,color='darkred',label='R')
plt.plot(s0[:,0],(s0[:,1]/s0[:,3]-ns0)/ns0,color='darkblue',label='S')
plt.semilogx()
plt.xlabel('radio [Mpc]')
plt.title('VOIDS BUENOS')
plt.legend()
```
