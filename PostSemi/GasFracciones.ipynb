```python
t=np.loadtxt('halosprop.dat')
```

```python
xbox=411.2170 
ybox=162.1655
zbox=453.0553
xc=413.621475-xbox+250
yc=162.604601-ybox+250
zc=448.953638-zbox+250
```

```python
d=np.sqrt((t[:,0]-xc)**2+(t[:,1]-yc)**2+(t[:,2]-zc)**2)
```

```python
np.max(d)
```

```python
cut0, = np.where((t[:,10]>0) & (t[:,5]>50))
cut1, = np.where((t[:,11]>0) & (t[:,5]>50))
cut2, = np.where((t[:,12]>0) & (t[:,5]>50))
cut3, = np.where((t[:,13]>0) & (t[:,5]>50))
len(cut0)
plt.plot(d[cut0],t[cut0,10],ls='none',marker='.',markersize=.1)
plt.plot(d[cut1],t[cut1,11],ls='none',marker='.',markersize=.1,color='blue')
plt.plot(d[cut2],t[cut2,12],ls='none',marker='.',markersize=.1)
plt.plot(d[cut3],t[cut3,13],ls='none',marker='.',markersize=.1,color='red')
```
