```python
t=np.loadtxt('halosprop_S.dat')
```

```python
#para ordenar ascendentemente segun la columna 0
s=t[np.argsort(t[:,0])]
```

```python
withst, = np.where(s[:,8]>0)
s=s[withst,:]
```

```python
rl=30
cut, = np.where(s[:,0]<rl)
print('halos en zona',len(cut))
print('cortar cada:',int(len(cut)/6))
```

```python
t1=s[:int(len(cut)/6),:]
t2=s[1*int(len(cut)/6):2*int(len(cut)/6),:]
t3=s[2*int(len(cut)/6):3*int(len(cut)/6),:]
t4=s[3*int(len(cut)/6):4*int(len(cut)/6),:]
t5=s[4*int(len(cut)/6):5*int(len(cut)/6),:]
t6=s[5*int(len(cut)/6):6*int(len(cut)/6),:]
```

```python
s[1*2644,0]
```

```python
masa_gas=1.81729961E-02
masa_dm=9.32880491E-02
```

```python
np.max((t1[:,9])/(t1[:,7]*masa_dm))
```
