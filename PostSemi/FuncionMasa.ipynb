```python
tr=np.loadtxt('/home/arodriguez/Void-s-Gastrophysics/PostSemi/halosprop2_R.dat')
ts=np.loadtxt('/home/arodriguez/Void-s-Gastrophysics/PostSemi/halosprop2_S.dat')
masa_dm=0.09328805
masa_gs=0.018172996
```

```python
#inr, = np.where(tr[:,0]<9.56)
#ins, = np.where(ts[:,0]<9.75)
#out,= np.where((ts[:,0]<30) & (ts[:,0]>25 ))

inr, = np.where(tr[:,0]<13)
ins, = np.where(ts[:,0]<10)
outs,= np.where((ts[:,0]<30) & (ts[:,0]>25 ))
outr,= np.where((tr[:,0]<30) & (tr[:,0]>25 ))
```

```python
rango=(10,np.max(tr[outr,5]))
bines0=16
minimo=10*masa_dm*1e10
maximo=np.max(ts[outs,5])*masa_dm*1e10
bines=np.logspace(np.log10(minimo),np.log10(maximo), bines0)
plt.title('funcion de particulas (masa)')

h1=tr[inr,5]*masa_dm*1e10
h2=ts[ins,5]*masa_dm*1e10
h3=np.append(tr[outr,5],ts[outs,5])*masa_dm*1e10

H1=plt.hist(h1,histtype='step',bins=bines,density='true',lw=2,color='blue',label='R dentro')
H2=plt.hist(h2,histtype='step',bins=bines,density='true',lw=2,color='red',label='S dentro')
H3=plt.hist(h3,histtype='step',bins=bines,density='true',lw=2,color='black',ls='--',label='fuera')
plt.legend()
plt.semilogy()
plt.semilogx()
```

```python
bincenter3  = []
bincenter2  = []
bincenter1  = []
for i in range(len(H3[0])):
    bincenter3.append((H3[1][i+1]-H3[1][i])*0.5 + H3[1][i])
    bincenter2.append((H2[1][i+1]-H2[1][i])*0.5 + H2[1][i])
    bincenter1.append((H1[1][i+1]-H1[1][i])*0.5 + H1[1][i])
```

```python
fig=plt.figure(figsize=(10,10))
fig.subplots_adjust(hspace=(0.3),bottom=0.12, top=0.98, left=0.17, right=0.97)

plt.plot(bincenter3,H3[0],label=('region externa'),color='black')
plt.plot(bincenter2,H2[0],label='void S',ls='dotted',color='red',lw=5)
plt.plot(bincenter1,H1[0],label='void R',ls='--',color='blue',lw=5)
plt.fill_between(bincenter3, 0, H3[0],color='grey', alpha=0.2)

plt.xticks([1,2],size=20)

plt.yticks(size=20)
plt.legend(fontsize='xx-large')
plt.xlabel('$M_{dm} [M_{\odot}]$',size=35)
plt.ylabel('$dN/DM\,[h/M_{\odot}]$',size=35)
#plt.ylim(0.002,6)
#plt.ticklabel_format(style='sci', axis='x')#, scilimits=(1,2))
plt.semilogy()
plt.semilogx()
plt.savefig(fname='/home/arodriguez/Void-s-Gastrophysics/SeleccionGraficos/FuncionesDeMasa.pdf',format='pdf')
```
