```python
%pylab inline
from math import acos
```

```python
t=np.loadtxt('halos_S1373.ascii',skiprows=19)
```

```python
xbox=403.8960
ybox=459.8882
zbox=440.9021

x=408.205481   
y=457.777839   
z=441.538681

x = x - xbox + 250
y = y - ybox + 250
z = z - zbox + 250
```

```python
d=np.sqrt((t[:,8]-x)**2 + (t[:,9]-y)**2 + (t[:,10]-z)**2)
dx=t[:,8]-x
dy=t[:,9]-y
dz=t[:,10]-z
lx=t[:,14]
ly=t[:,15]
lz=t[:,16]
spn=t[:,18]
```

```python
L = np.sqrt(lx**2+ly**2+lz**2)
mask1, = np.where(L > 0)
tita = (lx[mask1]*dx[mask1] + ly[mask1]*dy[mask1] + lz[mask1]*dz[mask1])/(L[mask1]*d[mask1])
#tita=np.arccos(tita)
#tita=tita*180./np.pi
```

```python
grandes, = np.where( t[mask1,1] > 500)
chicos, = np.where( t[mask1,1]<50)
invoid, = np.where(d[mask1] < 10)
outvoid, = np.where( d[mask1] > 20)
```

```python
plt.hist(tita,histtype='step',lw=3,color='black',bins=20,density='true',label='TODOS')
plt.hist(tita[grandes],histtype='step',linestyle='--',lw=2,color='red',density='true',label='grandes')
plt.hist(tita[chicos],histtype='step',linestyle='--',lw=2,color='blue',density='true',label='chicos')
plt.legend()
plt.xlabel('cos tita')
plt.title('Spin-flit')

```

```python
plt.hist(tita[invoid], histtype='step',color='black',lw=2,density='true',label='dentro')
plt.hist(tita[outvoid],histtype='step',color='cyan',lw=2,density='true',label='fuera')
plt.legend()
```
