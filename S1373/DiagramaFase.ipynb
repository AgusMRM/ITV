```python
t=np.loadtxt('diagfases_VOID.dat')
temp=np.log10(t[:,0])
dens=np.log10(t[:,1])
```

```python
plt.hexbin(dens,temp,cmap='viridis',bins='log')
plt.savefig(fname='/home/arodriguez/tesis/S1373/S1373_diagfaseVOID.pdf',format='pdf')
```

```python
DIF = np.where((dens < 2) & (temp < 5))
HALO = np.where((dens >= 2) & (temp < 5) & (t[:,2]==0))
WHIM = np.where((dens < 2) & (temp >= 5))
WCGM = np.where((dens >= 2) & (temp >= 5))
SF = np.where(t[:,2] > 0)
```

```python
print( 'particulas DIF', len(temp[DIF]))
print('particulas HALO', len(temp[HALO]))
print( 'particulas WHIM', len(temp[WHIM]))
print( 'particulas WCGM', len(temp[WCGM]))
print( 'particulas SF', len(temp[SF]))
```

```python
print('fraccion DIF', len(temp[DIF])/len(t))
print('fraccion HALO', len(temp[HALO])/len(t))
print('fraccion WHIM', len(temp[WHIM])/len(t))
print('fraccion WCGM', len(temp[WCGM])/len(t))
print('fraccion SF', len(temp[SF])/len(t))
```
