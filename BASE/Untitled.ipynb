```python
s=np.loadtxt('S1373part_base.dat')
r=np.loadtxt('R1198part_base.dat')
```

```python
rz=np.loadtxt('R1198_DMyTID.dat')
rt=np.loadtxt('R1198_TOTAL.dat')
rg=np.loadtxt('particlesprofile.dat')
```

```python
vol=(4./3.)*np.pi*(60**3)
mass_s=np.sum(s[:,1])
mass_r=np.sum(r[:,1])
rho_s=mass_s/vol
rho_r=mass_r/vol
```

```python
mass_rz=np.sum(rz[:,4])
mass_rg=np.sum(rg[:,4])
mass_rt=np.sum(rt[:,5])
rho_rz=mass_rz/vol
rho_rt=mass_rt/vol
rho_rg=mass_rg/vol
```

```python
plt.plot(s[:,0],(s[:,1]/s[:,3]-rho_s)/rho_s,label='BASE',color='blue',linestyle='--')
plt.semilogx()
plt.title('PERFIL PARTICULAS S1373')
plt.legend()
plt.xlabel('radio Mpc')
plt.ylabel('DELTA')
```

```python
plt.plot(r[:,0],(r[:,1]/r[:,3]-rho_r)/rho_r,label='BASE',color='blue',linestyle='--')
plt.plot(rz[:,0],(rz[:,4]/rz[:,3]-rho_rz)/rho_rz,label='ZOOM IN DM + TID',color='black')
plt.plot(rt[:,0],(rt[:,5]/rt[:,3]-rho_rt)/rho_rt,label='MASA TOTAL',color='green')
plt.plot(rg[:,0],(rg[:,4]/rg[:,3]-rho_rg)/rho_rg,label='MASA GAS',color='cyan')

plt.semilogx()
plt.title('PERFIL PARTICULAS R1198')
plt.legend()
plt.xlabel('radio Mpc')
plt.ylabel('DELTA')
```
